```python
# Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the 
#origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import seaborn as sns
# creating 2 random array
x=np.random.randn(100)
y=np.random.randn(100)

# creating dataframe to determine quadrants
df=pd.DataFrame({
    'x':x,
    'y':y
    
})
#defining the quadrants
def quadrant(row):
    if row['x'] >= 0 and row['y'] >=0:
        return 'Quadrant 1'
    elif row['x'] <= 0 and row['y'] >=0:
         return 'Quadrant 2'
    elif row['x'] <=0 and row['y']<=0:
        return 'Quadrant 3'
    else :
        return 'Quadrant 4' 
    
df['Quadrant']=df.apply(quadrant,axis=1)

#CREATING SCATTER PLOT
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='x',y='y',hue='Quadrant',palette=['blue','red','green','yellow'],s=100)

# add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'
plt.xlabel('x value')
plt.ylabel('y label')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.grid(True)
plt.show()
```

```python
#With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
from bokeh.plotting import figure,show
from bokeh.io import output_notebook

# Generate data for sine wave
x=np.linspace(0,4*np.pi,100)
y=np.sin(x)
# output
output_notebook()
# Create a new figure with a title and axis labels
p=figure(title='Sine Wave Function',y_axis_label='x', x_axis_label='x')

# adding linechart
p.line(x,y,legend_label='sine wave',line_width=2,color='blue')

# adding grid
p.grid.grid_line_color='gray'
p.grid.grid_line_alpha=0.5

# show
show(p)
```

```python
#Using Bokeh,generate a bar chat of randomly generated categorical data,color bars based on their values,add hover tooltips to display exact values,labelthe axes,and set the as 'Random Categorical Bar chats.'
##sir tried so many times but error doesn't fix
from bokeh.plotting import figure,show
from bokeh.models import HoverTool,ColumnDataSource
from bokeh.palettes import Spectral9
import numpy as np
import pandas as pd

#geneate random categorical data
categories=['a','b','c','d','e','f','g','h']
values=np.random.randint(1,100,size=len(categories))

# creating dataframe
df=pd.DataFrame({'categories':categories,
                 'values':values
})

#creating data colonm  source
source=ColumnDataSource(df)

#creating plot
p=figure(x_range=categories,plot_height=340,title='Random Categorical Bar Chart')

p.vbar(x='categories',top='values',width=0.9,color=[Spectral9[i%len(Spectral9)]for i in range(len(categories))],source=source)

#adding hover tool
hover=HoverTool(
    tooltips=[
        ('category','@categories'),
        ('value','@values')
        
    ]
)


p.add_tools(hover)

# adding axis labels
p.xaxis.axis_label='categoriess'
p.yaxis.axis_label='values'
# adding axis labels

show(p)
```
